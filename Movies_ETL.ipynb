```python
def clean_movie(movie):
    movie = dict(movie) #create copy of movie
    alt_titles = {}
    for key in ['Traditional','Mandarin','Also known as','Simplified','French','Hangul','Revised Romanization','McCune-Reischauer'
                 'Original title','Japanese','Hepburn','Literally','Cantonese','Chinese','Yiddish','Arabic','Romanized','Russian',
                 'Hebrew','Polish']:
        if key in movie:
            alt_titles[key] = movie[key]
            movie.pop(key)
    if len(alt_titles) > 0:
        movie['alt_titles'] = alt_titles
        
    return movie
```

```python
clean_movies = [clean_movie(movie) for movie in wiki_movies]
```

```python
wiki_movies_df = pd.DataFrame(clean_movies)
sorted(wiki_movies_df.columns.tolist())
```
